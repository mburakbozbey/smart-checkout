```python
import os
from glob import glob
import cv2
```

```python
classes = [path.split("\\")[-1] for path in glob("./EE492/dataset_492_vids_test/*")]

cdir = "./EE492/dataset_492_vids_test/"
wdir = "./EE492/dataset_492_test/"

for path in classes:
    vid_path = glob(cdir + path + "/*.mp4")
    for i,vid in enumerate(vid_path):
        vidcap = cv2.VideoCapture(vid)
        success,image = vidcap.read()
        count = 0
        write_path = wdir + path 
        while success:
            cv2.imwrite(write_path + "/vid" + str(i) + "_%d.jpg" % count, image)      
            success,image = vidcap.read()
            print('Read a new frame: ', success)
            count += 1
```
